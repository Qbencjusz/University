```python
from itertools import permutations
import numpy as np
import time
```

```python
bigram_dict = dict()
bigrams = []

with open('poleval_2grams.txt', 'r', encoding='utf8') as f:
    for line in f:
        l = line.split()
        if int(l[0]) > 3:
            bigrams.append(line)

for line in bigrams:
    l = line.split()
    if l[1] in bigram_dict:
        bigram_dict[l[1]].append((l[2], l[0]))
    else:
        bigram_dict[l[1]] = [(l[2], l[0])]
```

```python
def bi_score(perm):
    score = 0
    for i in range(len(perm)-1):
        if perm[i] in bigram_dict:
            for val in bigram_dict[perm[i]]:
                if val[0] == perm[i+1]:
                    score += int(val[1])
    return score
```

```python
def best_perms(ps):
    perms_len = len(ps)
    scores = np.zeros(perms_len)
    for i in range(perms_len):
        scores[i] = bi_score(ps[i])
    max_idxs = np.argsort(scores)[::-1]
    for i in range(3):
        print(ps[max_idxs[i]], scores[max_idxs[i]])
    correct_idx = np.where(max_idxs == 0)
    points = 1/(int(correct_idx[0]) + 1)
    print(f'points: {points}')
    return points
```

```python
sentences = []

with open('zad5test.txt', 'r', encoding='utf8') as f:
    for line in f:
        sentences.append(line.split())
```

```python
res = 0
t0 = time.time()
for sentence in sentences: 
    ps = list(permutations(sentence))
    res += best_perms(ps)
res /= len(sentences)
print(f'average: {res}')
t1 = time.time()
print(f'time: {t1 - t0} s')
```
